### Q4. make a new column which Consist of Sum of Income1 and Income2 

```python
df['Income.1']=df['Income.1'].str.replace("$","")
df
```

```python
df['Income.1']=df['Income.1'].replace("",np.nan)
```

```python
df
```

```python
df.info()
```

```python
df['Income.1']=df['Income.1'].astype(float)
df.info()
```

```python
df['Income.1'].fillna(df['Income.1'].mean(),inplace=True)
```

```python
df['Income.2'].fillna(df['Income.2'].mean(),inplace=True)
```

```python
df['Total Income']=np.ceil(df['Income.1']+df['Income.2'])
```
