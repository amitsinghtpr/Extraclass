## Question 3  fill Invalid Country name and replace nan value in country columns

```python
country_name= {'UAS': 'USA', 'NaN': 'Canada'}
df['Country'].replace(country_name, inplace=True)
df['Country'].fillna(method='ffill', inplace=True)
df
```

### Q4. make a new column which Consist of Sum of Income1 and Income2 

```python
df['Income.1'].fillna(0,inplace=True)
df['Income.2'].fillna(0,inplace=True)
df['Total Income']=pd.to_numeric(df['Income.1'].str.replace('[\$,]','',regex=True)).fillna(0) + df['Income.2'].fillna(0)
df
```
